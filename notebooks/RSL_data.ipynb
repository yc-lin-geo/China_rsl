```python
#load modules
import sys
sys.path.append('../')
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import PSTHM
import matplotlib.pyplot as plt
import pandas as pd

#set plotting style
%matplotlib inline
font = {'weight':'normal',
       'size':20}

matplotlib.rc('font',**font)
matplotlib.rcParams['figure.figsize'] = (12, 6)
import warnings
warnings.filterwarnings("ignore")
plt.rcParams["font.family"] = "Helvetica"
```

### Loading relative sea-level data used in this study and generate Figure 1b


```python
RSL_data = pd.read_excel('../data/Lin_et_al_China_RSL_data.xlsx',sheet_name=1)
RSL_age = RSL_data['Age (BP)']
RSL_age_2sigma = (RSL_data['Age +2s']+RSL_data['Age -2s'])/2
RSL_SLIPs = RSL_data['RSL (m)']
RSL_2sigma = (RSL_data['RSL +2s']+RSL_data['RSL -2s'])/2
```

```python
plt.figure(figsize=(12,12))
ax = plt.subplot(111)
PSTHM.plotting.plot_uncertainty_boxes(RSL_age,RSL_SLIPs, RSL_age_2sigma,RSL_2sigma,CE=False)
ax.set_xlim(11500,-100)
ax.set_ylim(-55,12)
```

```python

# Calculate histogram values
hist1, bins = np.histogram(RSL_age, bins=np.arange(12000,-100,-500)[::-1])
plt.figure(figsize=(12,4))
plt.bar(bins[:-1], hist1, width=np.diff(bins), align='edge',ec=(0.7,0,0,0.7),color=(1,0.82,0.86,0.15),linewidth=1.2)
plt.xlim(11500,-10)
```
